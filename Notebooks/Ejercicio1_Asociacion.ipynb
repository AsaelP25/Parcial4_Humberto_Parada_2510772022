```python
from mlxtend.frequent_patterns import apriori, association_rules
```

Verificar columnas del dataset

Preparar los datos para Apriori

```python
tabla_transacciones = df.groupby(["transaccion_id", "item"])["cantidad"].sum().unstack().fillna(0)

tabla_transacciones.head()
```

Convertir cantidades a 0 y 1

```python
tabla_transacciones = tabla_transacciones.applymap(lambda x: 1 if x > 0 else 0)

tabla_transacciones.head()
```

Se transformaron los datos para que cada fila representara una transacción y cada columna representara un producto.

Los valores se convirtieron a 1 y 0, donde 1 significa que el producto fue comprado en esa transacción y 0 significa que no fue comprado. Esta estructura es necesaria para aplicar el algoritmo Apriori.

Aplicar algoritmo Apriori

```python
itemsets_frecuentes = apriori(
    tabla_transacciones,
    min_support=0.03,
    use_colnames=True
)

itemsets_frecuentes.head()
```

Ordenar los itemsets frecuentes

```python
itemsets_frecuentes = itemsets_frecuentes.sort_values(by="support", ascending=False)

itemsets_frecuentes.head(10)
```

El soporte indica qué tan frecuente aparece un producto o combinación de productos dentro de todas las transacciones. Mientras más alto sea el soporte, más común es esa compra dentro del dataset.

Generar reglas de asociación

```python
reglas = association_rules(
    itemsets_frecuentes,
    metric="confidence",
    min_threshold=0.2
)

reglas.head()
```

Seleccionar columnas importantes

```python
reglas_importantes = reglas[
    ["antecedents", "consequents", "support", "confidence", "lift"]
]

reglas_importantes.head(10)
```

Ordenar las mejores reglas

```python
reglas_importantes = reglas_importantes.sort_values(by="lift", ascending=False)

reglas_importantes.head(10)
```

Mostrar las 10 reglas más relevantes

```python
top_10_reglas = reglas_importantes.head(10)

top_10_reglas
```

Interpretar reglas automáticamente

```python
for index, row in top_10_reglas.iterrows():
    antecedente = list(row["antecedents"])
    consecuente = list(row["consequents"])
    soporte = row["support"]
    confianza = row["confidence"]
    lift = row["lift"]

    print(f"Regla: Si compra {antecedente}, también compra {consecuente}")
    print(f"Soporte: {soporte:.2f}")
    print(f"Confianza: {confianza:.2f}")
    print(f"Lift: {lift:.2f}")
    print("-" * 60)
```
